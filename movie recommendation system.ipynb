```python
import numpy as np
import pandas as pd
import ast 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
movies.columns.values
movies = movies[['id','title','genres','keywords','overview','tagline']]
movies.head()
movies.isna().sum()
```

```python
credits.head()
credits.rename(columns={'movie_id':'id'},inplace=True)
credits.drop(columns='title',inplace =True)
credits.head()
```

```python
df = movies.merge(credits,on='id')
df
df.isna().sum()
```

```python
df.loc[df['tagline'].isna(),'tagline'] = ''
df.isna().sum()
```

```python
df.loc[df['overview'].isna(), 'overview'] = ''
df.isna().sum()
```

```python
df.head()
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return  (' ').join(L)
```

```python
df['genres'] = df['genres'].apply(convert)
df
```

```python
df['keywords'] = df['keywords'].apply(convert)
df
```

```python
def actors(obj):
        L =[]
        for i in ast.literal_eval(obj):
            if len(L) >=3:
                break
            else:
             L.append(i['name'])
        return ' '.join(L)

def director(obj):
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                return i['name']
        return ''
```

```python
df['cast'] = df['cast'].apply(actors)
df
```

```python
df['crew'] = df['crew'].apply(director)
df
```

```python
df['sumup'] = df.overview + df.tagline + df.genres +df.keywords + df.cast + df.crew
df
```

```python
df = df[['id','title','sumup']]
df
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return ' '.join(L)

df.sumup = df.sumup.apply(stem)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
skyline = CountVectorizer(max_features = 7000, stop_words= 'english' )
```

```python
horizon = skyline.fit_transform(df.sumup)
horizon = horizon.toarray()
horizon
```

```python
from sklearn.metrics.pairwise import cosine_similarity
celestial = cosine_similarity(horizon)
celestial.shape
```

```python
def recommend(movie):
    movie_index = df[df.title == movie].index[0]
    distances = celestial[movie_index]
    cosmic = sorted(list(enumerate(distances)), reverse= True, key = lambda x: x[1])
    return cosmic[1:6]
```

```python
azure = recommend('Newlyweds')
azure
for i in azure:
   print(df.iloc[i[0]].title)
```
